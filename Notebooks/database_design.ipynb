```python
import pandas as pd
import itertools
```

```python
df = pd.read_csv("../Data/Cleaned_Data/Olist_Brazillian_Ecommerce_Dataset_Cleaned.csv")
```

```python
df.info()
```

```python
date_cols = ['shipping_limit_date', 'order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in date_cols:
    df[col] = pd.to_datetime(df[col])
```

```python
df.info()
```

```python
print(df.columns)
```

# Normalizaing dataset

## Customers dataset

```python
customers = df[['customer_id', 'customer_unique_id', 'customer_zip_code_prefix', 'customer_city', 'customer_state']]
print(len(customers))
```

```python
customers = customers.drop_duplicates(subset = ['customer_id'])
print(len(customers))
```

## Products dataset

```python
products = df[['product_id', 'product_category_name', 'product_name_length', 'product_description_length', 'product_photos_qty', 'product_weight_g', 'product_length_cm', 'product_height_cm',
               'product_width_cm']]
print(len(products))
```

```python
products = products.drop_duplicates(subset = ['product_id'])
print(len(products))
```

## Sellers dataset

```python
sellers = df[['seller_id', 'seller_city', 'seller_state', 'seller_zip_code_prefix']]
print(len(sellers))
```

```python
sellers = sellers.drop_duplicates(subset = ['seller_id'])
print(len(sellers))
```

## Orders dataset

```python
orders = df[['order_id', 'customer_unique_id', 'order_status', 'order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']]
print(len(orders))
```

```python
orders = orders.drop_duplicates(subset = ['order_id'])
print(len(orders))
```

## Orders Duration dataset

```python
orders_duration = df[['order_id', 'apporval_time_mins', 'carrier_pickup_time_mins', 'shipping_time_mins', 'total_delivery_time_mins', 'order_delay_days', 'order_delay_time_mins']]
print(len(orders_duration))
```

```python
orders_duration = orders_duration.drop_duplicates(subset=['order_id'])
print(len(orders_duration))
```

## Order Items dataset

```python
order_items = df[['order_id', 'order_item_id', 'product_id', 'seller_id', 'shipping_limit_date', 'price', 'freight_value']]
print(len(order_items))
```

```python
order_items = order_items.drop_duplicates(subset = ['order_id', 'order_item_id'])
print(len(order_items))
```

## Payments dataset

```python
payments = df[['order_id', 'payment_sequential', 'payment_type', 'payment_installments', 'payment_value']]
print(len(payments))
```

```python
payments = payments.drop_duplicates(subset = ['order_id', 'payment_sequential'])
print(len(payments))
```

# Saving Normalized datasets

```python
customers.to_csv("../Data/Normalized_Data/customers.csv", index = False, encoding = "utf-8")
products.to_csv("../Data/Normalized_Data/products.csv", index = False, encoding = "utf-8")
sellers.to_csv("../Data/Normalized_Data/sellers.csv", index = False, encoding = "utf-8")
orders.to_csv("../Data/Normalized_Data/orders.csv", index = False, encoding = "utf-8")
orders_duration.to_csv("../Data/Normalized_Data/orders_duration.csv", index = False, encoding = "utf-8")
order_items.to_csv("../Data/Normalized_Data/order_items.csv", index = False, encoding = "utf-8")
payments.to_csv("../Data/Normalized_Data/payments.csv", index = False, encoding = "utf-8")
```
